```python
import pandas as pd
import json

video_data = pd.read_csv('C:/Users/Brooks Harris/Desktop/USvideos.csv')
video_data.head()
```

```python
# Loading category information
with open('C:/Users/Brooks Harris/Desktop/US_category_id.json', 'r') as file:
    categories = json.load(file)

# Creating a mapping of category IDs to category names
category_mapping = {int(category['id']): category['snippet']['title'] for category in categories['items']}
print(category_mapping)
```

```python
# Mapping the category names to the video data
video_data['category_name'] = video_data['category_id'].map(category_mapping)
video_data
```

```python
video_data['like_ratio'] = (video_data['likes'] / (video_data['likes'] + video_data['dislikes'])) * 100

profile_columns = ['video_id', 'title', 'channel_title', 'publish_time', 'trending_date', 'views', 'likes', 'dislikes', 'like_ratio', 'category_name']
profiles = video_data[profile_columns]

profiles
```

```python
import pandas as pd

# Assuming you have loaded your video data into the DataFrame `video_data`

# Convert 'publish_time' to datetime and ensure it's tz-naive
profiles['publish_time'] = pd.to_datetime(profiles['publish_time']).dt.tz_localize(None)

# Convert 'trending_date' using the correct format and ensure it's also tz-naive
profiles['trending_date'] = pd.to_datetime(profiles['trending_date'], format='%y.%d.%m').dt.tz_localize(None)

# Verify the conversions
print(profiles[['publish_time', 'trending_date']].info())
```

```python
# Now calculate the difference in days between trending date and publish date
profiles['days_till_viral'] = (profiles['trending_date'] - profiles['publish_time']).dt.days
```

```python
from sqlalchemy import create_engine, Column, Integer, String, Float, DateTime
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

# Define the base class
Base = declarative_base()

# Define a class corresponding to the 'profiles' table
class VideoProfile(Base):
    __tablename__ = 'profile_table'
    id = Column(Integer, primary_key=True)
    video_id = Column(String)
    title = Column(String)
    channel_title = Column(String)
    publish_time = Column(DateTime)
    views = Column(Integer)
    likes = Column(Integer)
    dislikes = Column(Integer)
    like_ratio = Column(Float)
    category_name = Column(String)
    days_till_viral = Column(Integer)

# Create an SQLite database and a table based on the class
engine = create_engine('sqlite:///youtube_video_profiles.db')
Base.metadata.create_all(engine)

# Establish a session to interact with the database
Session = sessionmaker(bind=engine)
session = Session()

# Iterate through each row in your DataFrame and insert it into the database
for index, row in profiles.iterrows():
    video_profile = VideoProfile(
        video_id=row['video_id'],
        title=row['title'],
        channel_title=row['channel_title'],
        publish_time=row['publish_time'],
        views=row['views'],
        likes=row['likes'],
        dislikes=row['dislikes'],
        like_ratio=row['like_ratio'],
        category_name=row['category_name'],
        days_till_viral=row['days_till_viral']
    )
    session.add(video_profile)

# Commit the session to save the data to the database
session.commit()

# Close the session
session.close()
```
